```python
%load_ext autoreload
%autoreload 2

import os, sys
os.chdir("/mnt/arrakis/sietch/Resume")
sys.path.append("./src")
```

```python
from resumeai.models import Resume
from resumeai.utils import extract_object

from jinja2 import Environment, FileSystemLoader
from rich import print
```

```python
with open('./profile.md') as f:
    text = ''.join(f.readlines())

resume = extract_object(
    prompt="Extract the user's resume from the text below:",
    text=text,
    cls=Resume,
)
```

```python
import subprocess
import shutil
from pathlib import Path
```

```python
shutil.rmtree("./rendered", ignore_errors=True)


template_dir = './templates/TLCresume'

env = Environment(loader=FileSystemLoader(template_dir))

rendered = "./rendered"
os.makedirs(f"{rendered}/sections", exist_ok=True)

shutil.copyfile(f"{template_dir}/TLCresume.sty", f"{rendered}/TLCresume.sty")
shutil.copyfile(f"{template_dir}/_header.tex", f"{rendered}/_header.tex")

for section in resume:
    template = env.get_template(f"{section.template}.tex.j2")
    text = template.render(items=section.get_items())
    
    with open(f"{rendered}/{section.template}.tex", "w") as f:
        f.write(text)

shutil.make_archive(rendered, 'zip', rendered)

subprocess.run("pdflatex -interaction=nonstopmode resume.tex", shell=True, cwd=rendered)
```
